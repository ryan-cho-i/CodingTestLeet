```python
def minArea(points) :

    minX = float('inf')
    maxX = -float('inf')
    minY = float('inf')
    maxY = -float('inf')

    for x, y in points:
        minX= min(x, minX)
        maxX= max(x, maxX)
        minY= min(y, minY)
        maxY= max(x, maxY)
    
    return (max(maxX-minX+2, maxY-minY+2)) ** 2
```
